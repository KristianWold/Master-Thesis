### Generating data

```python
np.random.seed(42)
x = np.random.uniform(0, np.pi, (100,1))
y = np.sin(3*x) + np.sin(7.7*x) - np.sin(11.3*x) - np.sin(16.7*x) + np.sin(20.1*x)
y = y - np.min(y)
y = y/np.max(y)
```

```python
plt.plot(x,y, "o")
plt.show()
```

```python
np.random.seed(42)
optimizer = Adam(lr=1e-2)

layer1 = Dense(n_features=1, n_targets=5, scale = 1, activation = Sigmoid())
layer2 = Dense(n_features=5, n_targets=5, scale = 1, activation = Sigmoid())
layer3 = Dense(n_features=5, n_targets=1, scale = 1, activation = Identity())
layers =[layer1, layer2, layer3]
network = NeuralNetwork(layers, optimizer)
```

```python
network.train(x,y, epochs = 10000, verbose=True)
```

```python
np.random.seed(42)
backend = Aer.get_backend('qasm_simulator')

layer1 = QLayer(n_qubits=1, n_features=1, n_targets=3, encoder=Encoder(), ansatz=Ansatz(), reps=1, scale=2*np.pi, backend=backend, shots=10000)
layer2 = QLayer(n_qubits=3, n_features=3, n_targets=3, encoder=Encoder(), ansatz=Ansatz(), reps=1, scale=2*np.pi, backend=backend, shots=10000)
layer3 = QLayer(n_qubits=3, n_features=3, n_targets=1, encoder=Encoder(), ansatz=Ansatz(), reps=1, scale=1, backend=backend, shots=10000)


layers = [layer1, layer2, layer3]

optimizer = Adam(lr=0.1)
network = NeuralNetwork(layers, optimizer)
```

```python
network.train(x,y, epochs = 100, verbose=True)
```
