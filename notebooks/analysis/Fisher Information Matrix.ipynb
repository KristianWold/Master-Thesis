# Fisher Information Matrix, Analysis

```python
fim1 = loader(data_path("fim_QNN_depth_4_width_5_reps_1_shots_100k"))
eigen1 = fim1.eigen()
fr1 = fim1.fisher_rao()

fim2 = loader(data_path("fim_QNN_depth_2_width_5_reps_2_shots_100k"))
eigen2 = fim2.eigen()
fr2 = fim2.fisher_rao()

fim3 = loader(data_path("fim_DNN_depth_4_width_5"))
eigen3 = fim3.eigen()
fr3 = fim3.fisher_rao()
```

```python
fig=plt.figure(figsize=(6,4), dpi = 130, facecolor='w', edgecolor='k')
plt.plot(np.log10(eigen1), "o")
plt.plot(np.log10(eigen2), "o")
plt.plot(np.log10(eigen3), "o")
plt.xlabel("Index")
plt.ylabel("log10 Eigenvalue")
plt.legend(["QKN with L=4, d=5, reps=1", "QKN with L=2, d=5, reps=2", "DNN with L=4, d=5"])
plt.grid()

plt.show()
fig.savefig(image_path("FIM_layer_2.pdf"), bbox_inches='tight')
```
