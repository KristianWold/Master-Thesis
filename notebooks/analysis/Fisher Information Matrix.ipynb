# Fisher Information Matrix, Analysis

## 4 qubits

```python
fim1 = loader(data_path("fim_QCN_depth_1_width_4_reps_18"))
eigen1 = np.mean(np.array([fim.eigen() for fim in fim1]), axis=0) 

fim2 = loader(data_path("fim_QCN_depth_3_width_4_reps_2"))
eigen2 = np.mean(np.array([fim.eigen() for fim in fim2]), axis=0) 

fim3 = loader(data_path("fim_DNN_depth_5_width_4"))
eigen3 = np.mean(np.array([fim.eigen() for fim in fim3]), axis=0) 
```

```python
fig=plt.figure(figsize=(6,4), dpi = 130, facecolor='w', edgecolor='k')
plt.plot(np.log10(eigen1)[:60], "o", markersize=4)
plt.plot(np.log10(eigen2)[:60], "o", markersize=4)
plt.plot(np.log10(eigen3)[:60], "o", markersize=4)
plt.xlabel("Index")
plt.ylabel("log10 Eigenvalue")
plt.legend(["(A) QNN, reps=18", "(C) QCN, reps=2, L=3, d=4", "(E) DNN, L=5, d=4"])
plt.grid()


print(eigen1.shape, eigen2.shape, eigen3.shape)
plt.show()
fig.savefig(image_path("FIM_qubits_4.pdf"), bbox_inches='tight')
```
