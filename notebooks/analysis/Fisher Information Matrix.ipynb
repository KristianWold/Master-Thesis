# Fisher Information Matrix, Analysis

```python
fim1 = loader(data_path("fim_QKN_depth_1_width_4_reps_6_shots_100k"))
eigen1 = np.mean(np.array([fim.eigen() for fim in fim1]), axis=0) 

fim2 = loader(data_path("fim_QKN_depth_3_width_4_reps_2_shots_100k"))
eigen2 = np.mean(np.array([fim.eigen() for fim in fim2]), axis=0) 

fim3 = loader(data_path("fim_DNN_depth_6_width_4"))
eigen3 = np.mean(np.array([fim.eigen() for fim in fim3]), axis=0) 
```

```python
fig=plt.figure(figsize=(6,4), dpi = 130, facecolor='w', edgecolor='k')
plt.plot(np.log10(eigen1), "o", markersize=4)
plt.plot(np.log10(eigen2), "o", markersize=4)
plt.plot(np.log10(eigen3), "o", markersize=4)
plt.xlabel("Index")
plt.ylabel("log10 Eigenvalue")
plt.legend(["QKN, L=1, d=4, reps=6", "QKN, L=3, d=4, reps=2", "DNN, L=6, d=4"])
plt.grid()

plt.show()
fig.savefig(image_path("FIM_qubits_4_depth_4.pdf"), bbox_inches='tight')
```
